# Normalize ChIPseq data with ChIP-Rx

```python
# normalization factors

# to get the normalization factors check Quantitative ChIP-Seq Normalization Reveals Global
Modulation of the Epigenome by Orlando et al. 
dmso_factor = 4.11
a485_factor = 4.68
gne781_factor = 3.33
ag_factor = 4.95
rv62_factor = 2.61
```

```python
import pandas as pd

dmso = pd.read_csv('dmso.bedgraph', sep='\t', header=None)

dmso
```

```python
dmso[3] = dmso[3].apply(lambda x: x * dmso_factor)
```

```python
dmso
```

```python
a485 = pd.read_csv('a485.bedgraph', sep='\t', header=None)

a485
```

```python
a485[3] = a485[3].apply(lambda x: x * a485_factor)
a485
```

```python
gne = pd.read_csv('gne781.bedgraph', sep='\t', header=None)

gne
```

```python
gne[3] = gne[3].apply(lambda x: x * gne781_factor)
gne
```

```python
ag = pd.read_csv('ag.bedgraph', sep='\t', header=None)

ag
```

```python
ag[3] = ag[3].apply(lambda x: x * ag_factor)
ag
```

```python
rv62 = pd.read_csv('rv62.bedgraph', sep='\t', header=None)

rv62
```

```python
rv62[3] = rv62[3].apply(lambda x: x * rv62_factor)

rv62
```

```python
dmso.to_csv(r'ndmso.bedgraph', header=None, index=None, sep='\t')
```

```python
a485.to_csv(r'na485.bedgraph', header=None, index=None, sep='\t')
gne.to_csv(r'ngne.bedgraph', header=None, index=None, sep='\t')
ag.to_csv(r'nag.bedgraph', header=None, index=None, sep='\t')
rv62.to_csv(r'nrv62.bedgraph', header=None, index=None, sep='\t')
```
